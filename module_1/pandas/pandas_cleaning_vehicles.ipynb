After really thorough analysis, I concluded that columns with more than 30% NaNs are useless and should thus be dropped.

```python
# get a list of columns, where relative frequency of NaNs > 30%
null_perc = null_cols[null_cols > 0] / len(df) 
to_drop = null_perc[null_perc > .3].index
to_drop
```

# 2. Information contained in the columns
(Cardinality of columns)

```python
df["barrels08"].head()
```

```python
df["barrels08"].min()
```

```python
np.percentile(df["barrels08"], 90)
```

```python
def detect_low_variance(df, column, perc):
    min_value = df[column].min()
    perc_value = np.percentile(df[column], perc)
    return min_value == perc_value
```

```python
detect_low_variance(df, "barrels08", 90)
```

```python
df.select_dtypes(include=[np.number]).columns
```

```python
low_variance = [col for col in df.select_dtypes(include=[np.number]).columns if detect_low_variance(df, col, 90)]
```

```python
df = df.drop(low_variance, axis=1)
```

# 3. Outliers

Using IQR

```python
df_desc = df.describe().transpose()
df_desc = df_desc.assign(IQR=df_desc["75%"] - df_desc["25%"])
```

```python
df_desc
```

```python
df_desc.index
```

```python
outliers = pd.DataFrame(columns=df.columns)

for col in df_desc.index:
    iqr = df_desc.loc[col, "IQR"]
    cutoff = iqr * 6
    lower_threshold = df_desc.loc[col, "25%"] - cutoff
    upper_threshold = df_desc.loc[col, "75%"] + cutoff
    results = df[(df[col] < lower_threshold) | (df[col] > upper_threshold)].copy()
    results = results.assign(outlier=col)
```
